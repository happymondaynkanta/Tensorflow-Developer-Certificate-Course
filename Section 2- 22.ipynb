Manipulating tensors (tensor operations)
Finding patterns in tensors (numberical representation of data) requires manipulating them.

Basic operations
You can perform many of the basic mathematical operations directly on tensors using Pyhton operators such as, +, -, *,/.

```python
# You can add values to a tensor using the addition operator

import tensorflow as tf
tensor = tf.constant([[10, 7], [3, 4]])
tensor + 10  #tensor = tensor + 10
```

```python
# Remember that the Original tensor is unchanged
tensor
```

```python
# Multiplication (known as element-wise multiplication)
tensor * 10
```

```python
# Subtraction
tensor - 10
```

```python
# Use the tensorflow built-in function equivalent of the '*' (multiply) operator
# The buit-in operation of tensor speedily up more faster while using GPU.
tf.multiply(tensor, 10)
```

```python
# The original tensor is still unchanged
tensor
```

**Section 2-23: Matrix multiplication with tensor part 1**

One of the most common operations in machine learning algorithms is matrix multiplication.

TensorFlow implements this matrix multiplication functionality in the tf.matmul() method.

The main two rules for matrix multiplication to remember are:

Rule 1: The inner dimensions must match:

(3, 5) @ (3, 5) won't work\
(5, 3) @ (3, 5) will work\
(3, 5) @ (5, 3) will work\

Rule 2:The resulting matrix has the shape of the outer dimensions:

(5, 3) @ (3, 5) -> (5, 5)\
(3, 5) @ (5, 3) -> (3, 3)\
ðŸ”‘ Note: '@' in Python is the symbol for matrix multiplication.

```python
# Matrix multiplication in TensorFlow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix multiplication with Python operator '@'
tensor @ tensor
```

```python
# Create (3, 2) tensor
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])

# Create another (3, 2) tensor
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])
X, Y
```

```python
# Try to matrix multiply them (will error) because the 3 * 2 is not equal to 3*2 matrrix form (different shapes)
X @ Y
```

Trying to matrix multiply two tensors with the shape (3, 2) errors because the inner dimensions don't match.

We need to either:

Reshape X to (2, 3) so it's (2, 3) @ (3, 2).

Reshape Y to (3, 2) so it's (3, 2) @ (2, 3).

We can do this with either:

tf.reshape() - allows us to reshape a tensor into a defined shape.
tf.transpose() - switches the dimensions of a given tensor.

```python
# Let's try tf.reshape() first.

# Example of reshape (3, 2) -> (2, 3)
tf.reshape(Y, shape=(2, 3))
```
